```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#plot day and temp
day=[1,2,3,4,5,6,7]
temp=[50,51,52,48,47,49,46]
```

```python
plt.plot(day,temp)
```

```python
plt.plot(day,temp,color='green')
```

```python
plt.xlabel('day')
plt.ylabel('temp')
plt.plot(day,temp,color='green',linestyle='dashed',marker='o')
```

```python
#Legend ,Grid
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")
plt.legend(loc='best')
```

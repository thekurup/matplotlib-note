```python
#Legend ,Grid
#generate maximum ,minimum and average temperature report
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")
plt.grid()
plt.legend()
```

```python
#create a histogram by using an random  variable array in which itz mean =100,sd=20,and range = 200
np.random.seed(10)
#step1 creating an normal distrubution random array in data
data=np.random.normal(100,20,200)
#step2 plotting it in histogram
plt.hist(data,rwidth=0.8)
print(data.mean())
```

```python
#step3 plotting it in box plot
plt.boxplot(data)
#increasing the figures size
plt.figure(figsize=(10,5))
plt.show()
```

```python
#ploting multiple array in a one boxplot

#step 1 ,creating multiple array and storing in
np.random.seed(10)
data1=np.random.normal(100,10,200)
data2=np.random.normal(100,20,200)
data3=np.random.normal(100,30,200)
data4=np.random.normal(100,40,200)

#step2 storing all data in on variable

data=[data1,data2,data3,data4]

#step3 creating boxplot size and print boxplot

plt.figure(figsize=(10,5))
plt.boxplot(data)
plt.show()#removing unwanted writing
print("mean of data1",data1.mean())
print("mean of data2",data2.mean())
print("mean of data3",data3.mean())
print("mean of data4",data4.mean())
```

```python
#vertical way of printing boxplot
plt.boxplot(data,vert=False,patch_artist=True)
plt.show()
```

```python
#PIE PLOT
 #create a pie plot home details
    #step1 setting up values
values = [1400,600,300,410,250]
data = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]

# step2 giving the pie size
plt.figure(figsize=(11,7))
plt.axis=('equal')
#axis and figsize always come in top of plotting

#step 3  ploting it in pie and incresing radius to show percentage inside it

plt.pie(values,labels=data,radius=1.2,autopct="%0.2ff%%",explode=[0.1,0,0,0.2,0],startangle=90)
#explode for cutting the pie and autocat for providing float values
plt.legend()
# plt.show()
#saving pie chart
plt.savefig("piechart.jpg")
```
